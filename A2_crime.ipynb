Ann Arbor has public data for crimes committed along with time stamps and locations. These data were retrieved from https://www.a2gov.org/services/data/Pages/default.aspx, provided by the Ann Arbor Police Department.

````python
data = pd.read_csv('PartAB Crime Offenses.csv',na_values='None')
data['POINT_X'] = pd.to_numeric(data['POINT_X'])
data['POINT_Y'] = pd.to_numeric(data['POINT_Y'])
data['DATE'] = pd.to_datetime(data['DATE'])
````

## Plotting locations of crimes in Ann Arbor using Google Maps API

I want to know where the parts in A2 where all the crimes are being committed. I used the ```bokeh``` package, which is an alternative plotting software for ```python``` which allows for Google Maps to be imported as a backend for plotting geographical data.

I was following this blog:
http://www.bigendiandata.com/2017-06-27-Mapping_in_Jupyter/

````python
from bokeh.models import (
    GMapPlot, GMapOptions, Range1d, Circle, ColumnDataSource, PanTool, 
    WheelZoomTool, BoxSelectTool, HoverTool,CategoricalColorMapper
)
from bokeh.io import show, output_notebook
````

````python
%matplotlib inline
map_options = GMapOptions(lat=42.281475, lng=-83.748523,map_type='roadmap',zoom=13)
plot = GMapPlot(x_range=Range1d(),y_range=Range1d(),map_options=map_options,plot_width=1000,plot_height=1000)

# For GMaps to function, Google requires you obtain and enable an API key:
#
#     https://developers.google.com/maps/documentation/javascript/get-api-key
#
# Replace the value below with your personal API key:
plot.api_key = 'AIzaSyDioCR8a_McT9zOJGGRqT947gKRfiPA8Pk'

# let's only plot the crimes committed for the year 2017
subset = data[(data['DATE'] >= '1/1/2017') & (data['DATE'] < '1/1/2018')]
colors = subset['CRIME_PART'].apply(lambda x: 'red' if x=='A' else 'blue').tolist()

source = ColumnDataSource(
    data=dict(
        lat=subset['POINT_Y'].tolist(),
        lon=subset['POINT_X'].tolist(),
        date=subset['DATE'].dt.strftime('%m-%d-%Y %H:%M').tolist(),
        location=subset['ADDRESS'].tolist(),
        crime=subset['CRIME_DESC'].tolist(),
        charge=subset['CHARGE_DESC'].tolist(),
        part=subset['CRIME_PART'].tolist()
    )
)

# let's color the points based on the crime part
color_mapper = CategoricalColorMapper(factors=['A','B'], palette=['red','blue'])
circle = Circle(x="lon", y="lat", size=5, fill_color={"field":"part","transform":color_mapper},
                fill_alpha=0.2, line_color=None)

plot.add_glyph(source,circle)

plot.add_tools(PanTool(), WheelZoomTool(), BoxSelectTool(), HoverTool())

# allow information to pop up when you hover your mouse over the crime position
hover = plot.select(dict(type=HoverTool))
hover.tooltips = [
    ('Date','@date'),
    ('Location','@location'),
    ('Crime','@crime'),
    ('Charge','@charge')
]

output_notebook()
show(plot)
````

Downtown has the highest concentration of Part A crimes: a lot of shoplifting, assault, thefts, retail fraud, DUIs. Also, having seen this plot before, I'm also going to do something crazy and plot the temperature of over that time frame as well. I requested data from 4 NOAA weather stations located around Ann Arbor. https://www.ncdc.noaa.gov/cdo-web/search

````python
wdf = pd.read_csv('1231326.csv')
wdf['DATE'] = pd.to_datetime(wdf['DATE'])
wdf.head()
````

Let's plot the number of crimes per week. I didn't include 2012 because it looks like this public data set wasn't really being used much the first year, so I don't think it is representative.

````python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

from matplotlib import rcParams
rcParams['font.size'] = 16

# group by week and crime part, then plot each crime part separately
perweek=data['CASE_NUM'].groupby([data['DATE'].dt.to_period("W"),data['CRIME_PART']]).count().reset_index()
fig,ax=plt.subplots(figsize=(15,6))
for key,grp in perweek.groupby('CRIME_PART'):
    if key == 'A':
        color = 'red'
    else:
        color = 'blue'
    ax = grp.plot(ax=ax, kind='line', x='DATE', y='CASE_NUM', c=color, label='Part {}'.format(key))

ax.set_xlabel('Date',size=16)
ax.set_ylabel('Number of crimes per week')
ax.set_xlim('1/1/2013','1/1/2018')
ax.set_ylim(0,250)

# plotting home games
handles, labels = ax.get_legend_handles_labels()
for h in homegames:
    hline = ax.axvline(h,color='black',alpha=0.7,linestyle=':')
handles.append(hline)
labels.append('Michigan Football Home Games')

# plotting other sports
for s in sports:
    sline = ax.axvline(s,color='purple',alpha=0.7,linestyle=':')
handles.append(sline)
labels.append('Other sports at Michigan Stadium')

# plotting art fair
for a in artfair:
    aline = ax.axvline(a,color='orange',alpha=0.7,linestyle=':')
handles.append(aline)
labels.append('Ann Arbor Art Fair')

# plotting hash bash
for b in hashbash:
    bline = ax.axvline(b,color='green',alpha=0.7,linestyle=':')
handles.append(bline)
labels.append('Hash Bash')
    
# overlay temperature averaged over all weather stations for a given week
daily = wdf[['TMAX','TMIN']].groupby(wdf['DATE'].dt.to_period("W")).mean().reset_index()
offset = 30
ax.fill_between(daily["DATE"],daily['TMIN']+offset,daily['TMAX']+offset,color='grey',alpha=0.3)
ax2 = ax.twinx()
ax2.set_ylim(ax.get_ylim()[0]-offset,ax.get_ylim()[1]-offset)

ax2.yaxis.set_ticks(np.arange(7)*20-20)
ax2.set_ylabel('Average Weekly'+' '*25+'\nTemperature ($^\circ$F)'+' '*25)
    
ax.legend(handles,labels,fontsize=12,loc='upper center',framealpha=1)
ax.set_title('Crime rates in Ann Arbor, MI')
    
plt.show()
````

This plot seems to suggest that Michigan football fans are really rowdy. There are really high spikes in Part B crimes occuring on the days when Michigan home football games occur. This could be the result of there being more people in town; however, we didn't see this trend occur when other sports games were being played at Michigan Stadium. There also seems to be somewhat of a cyclical behavior in Part A crimes which seems to peak in the summer and drop in the winter. It looks like the data match really well to the number of crimes. I have a feeling people don't want to be outside in the winter and are thus less likely to get in trouble with the police. Below is the correlation. It isn't super strong, but by eye it looks really good.

````python
x = data[(data['CRIME_PART'] == 'A') & 
            (data['DATE'] >= '1/1/2014') & 
            (data['DATE'] < '1/1/2018')]
x = x['CASE_NUM'].groupby(x['DATE'].dt.to_period("W")).count()
y = wdf[(wdf['DATE'] >= '1/1/2014') & 
       (wdf['DATE'] < '1/1/2018')]
y = y[['TMIN','TMAX']].groupby(y['DATE'].dt.to_period("W")).mean()
z = y.join(x)
z.corr()
````
